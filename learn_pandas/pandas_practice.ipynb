```python
data[data['Age'] < 20].head()
```

```python
data[(data['Age'] < 20) & (data['Sex']=='female')].head()
```

### 增删改

```python
dic = {'PassengerId':[0,1],'Survived':[0,1],'Pclass':[2,2],'Name':['Tom','Jerry'],'Sex':['female','male'],'Age':[20.0,12.0],'SibSp':[1,2]}
data2 = pd.DataFrame(dic)
data2
```

### 行整合

```python
concated = pd.concat([data,data2])
concated.tail()
```

```python
data.append(data2, ignore_index=True).tail() # ignore_index=Truew
```

### 列整合

```python
cabin_dummies = pd.get_dummies(data['Cabin'], prefix='Cabin') # one-hot 编码
cabin_dummies.head()
```

```python
pd.concat([data, cabin_dummies], axis=1).head()
```

### 聚合操作

如果不对原始数据作限制的话，聚合函数会自动选择数值型数据进行聚合计算。

```python
data.groupby(['Sex','Pclass']).mean()
```

还可以对每个分组计算多个统计量

```python
data.groupby(['Sex','Pclass']).mean().loc['female',2]
```
